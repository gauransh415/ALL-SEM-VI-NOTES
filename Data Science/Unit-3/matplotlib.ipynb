```python
from scipy.stats import norm

# Data 
data = np.random.randn(10000)  # 1000 random values from a normal distribution 

# Histogram 
plt.hist(data, bins=50, color='green', edgecolor='black', density=True)  # Use density=True for PDF

# Generate points for the normal distribution curve
xmin, xmax = plt.xlim()  # Get the range of x-axis
x = np.linspace(xmin, xmax, 100)  # Generate 100 points between xmin and xmax
p = norm.pdf(x, np.mean(data), np.std(data))  # Calculate PDF

# Plot the normal distribution curve
plt.plot(x, p, 'r-', linewidth=2, label='Normal Dist Curve')

# Title and Labels 
plt.title("Histogram of Random Data with Normal Distribution Curve") 
plt.xlabel("Value") 
plt.ylabel("Density") 

# Add legend
plt.legend()

# Show plot 
plt.show() 
```

The string `'%1.1f%%'` is a format string used for percentage formatting in Python, often used in conjunction with the `autopct` parameter of the `pie` function in Matplotlib. Let's break it down:

*   `%`: This signifies the start of a replacement field in the format string.
*   `1.1f`: This is the format specifier.
    *   The first `1` specifies the minimum width of the entire number, including the decimal point and any digits before it. If the number requires more space than this, the width will be ignored.
    *   The second `1` (after the `.`) specifies the number of digits to display after the decimal point.
    *   `f` indicates that the number should be formatted as a floating-point number.
*   `%%`: This represents a literal percent sign (`%`) in the output string. Since `%` has a special meaning in format strings, you need to escape it by using `%%` to display it as a character.

In the context of `matplotlib.pyplot.pie`, setting `autopct='%1.1f%%'` will format the percentage labels on the pie chart slices to show one digit after the decimal point. For example, a slice representing 25.37% of the data would be labeled as "25.4%".


```python
# Data 
labels = ['A', 'B', 'C', 'D'] 
sizes = [15, 30, 45, 10] 
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'] 

# Pie chart 
plt.pie(
    sizes, labels=labels, colors=colors, 
    autopct='%1.1f%%', 
    startangle=90   # anticlockwise sequencially
) 

# Title 
plt.title("Category Distribution") 
# Show plot 
plt.show() 
```

```python
x = [1, 2, 3, 4, 5] 
y = [10, 20, 30, 40, 50]

# Creating 2x2 grid 
fig, axs = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(10, 8))
plt.subplots_adjust(wspace=0.3, hspace=0.4)  # Adjust the spacing between subplots 

# Plot 1: Line plot 
axs[0, 0].plot(x, y, color='blue') 
axs[0, 0].set_title("Line Plot") 

# Plot 2: Bar plot 
axs[0, 1].bar(x, y, color='orange') 
axs[0, 1].set_title("Bar Plot") 

# Plot 3: Scatter plot 
axs[1, 0].scatter(x, y, color='red') 
axs[1, 0].set_title("Scatter Plot") 

# Plot 4: Histogram 
import numpy as np 
data = np.random.randn(1000) 
axs[1, 1].hist(data, bins=30, color='green', edgecolor='black') 
axs[1, 1].set_title("Histogram") 

# Adjust layout 
plt.tight_layout() 

# Show the plots 
plt.show() 
```

```python
# Sample Data 
x = [1, 2, 3, 4, 5] 
y = [10, 20, 25, 30, 40] 

# Create a plot 
plt.plot(x, y, marker='o', markersize=10, linestyle='-', linewidth=2, color='blue')

# Set major and minor ticks 
plt.xticks(
    x, 
    ['One', 'Two', 'Three', 'Four', 'Five'],   # Custom tick labels for x-axis 
    fontsize=12,
    rotation=45
)  
plt.yticks(
    [10, 25, 40], 
    ['Low', 'Medium', 'High'],  # Custom tick labels for y-axis 
    fontsize=10, 
    rotation=90
)

# Display plot 
plt.show() 
```

```python
from matplotlib.ticker import MaxNLocator 

# Create plot
plt.plot(x, y, marker='o', markersize=10, linestyle='-', linewidth=2, color='blue')

# Automatically set 5 major ticks on the y-axis 
plt.gca().yaxis.set_major_locator(MaxNLocator(nbins=10)) 
plt.gca().xaxis.set_major_locator(MaxNLocator(nbins=5))

plt.grid(which='major', linestyle=':', linewidth=0.5)  # Dotted gridlines for minor ticks 
plt.show() 
```

```python
from matplotlib.ticker import AutoMinorLocator 

plt.plot(x, y, marker='o', markersize=10, linestyle='-', linewidth=2, color='blue')

# Enabling minor ticks 
plt.minorticks_on() 

# Customizing minor ticks
plt.gca().xaxis.set_minor_locator(AutoMinorLocator(4))  # 4 minor ticks between each major tick 
plt.gca().yaxis.set_minor_locator(AutoMinorLocator(5))  # 5 minor ticks between each major tick 

# Adding gridlines for minor ticks 
plt.grid(which='minor', linestyle=':', linewidth=0.5)  # Dotted gridlines for minor ticks 
plt.show() 
```

```python
from matplotlib.ticker import FuncFormatter 

# Custom function to format ticks 
def currency_formatter(x, pos): 
    return f"${x:.2f}"  # Format as currency 

plt.plot(x, y) 

# Apply custom tick formatter to y-axis 
plt.gca().yaxis.set_major_formatter(FuncFormatter(currency_formatter)) 
plt.show() 
```

```python
# Sample data 
x = [1, 2, 3, 4, 5] 
y = [10, 20, 25, 30, 40] 

plt.plot(x, y, label="Sales Growth", color='blue') 

"""
 xy=(5, 40): The coordinates where the annotation arrow will point. 
 xytext=(4, 35): The position where the annotation text will appear. 
 arrowprops=dict(facecolor='red', arrowstyle='->'): Specifies the appearance of the arrow 
(red color and a simple arrow). 
 fontsize=12: Font size of the annotation text. 
 color='black': Color of the text. 
"""

# Annotating a specific point 
plt.annotate(
    text='Highest Point',
    xy=(5, 40), 
    xytext=(4, 35), 
    arrowprops=dict(
        facecolor='#000000', 
    ), 
    fontsize=12, 
    color='black'
)

# Labels and title 
plt.xlabel("Time (Months)") 
plt.ylabel("Sales ($)") 
plt.title("Sales Growth with Annotation") 
plt.legend() 
plt.show()
```

```python
# Draw a vertical line at x=3 
plt.axvline(x=3, color='red', linestyle='--', linewidth=2) 

# Draw a horizontal line at y=20 
plt.axhline(y=20, color='green', linestyle='-', linewidth=2) 

# Labels and title 
plt.xlabel("Time (Months)") 
plt.ylabel("Sales ($)") 
plt.title("Sales Growth with Lines") 
plt.show() 
```

```python
import matplotlib.patches as patches 

# Plot the data 
plt.plot(x, y, label="Sales Growth", color='blue') 

# Create a rectangle 
rect = patches.Rectangle(xy=(1.5, 15), width=2, height=15, lw=5, edgecolor='purple', facecolor='orange', alpha=0.3) 
plt.gca().add_patch(rect) 

# Labels and title 
plt.xlabel("Time (Months)") 
plt.ylabel("Sales ($)") 
plt.title("Sales Growth with Rectangle") 
plt.legend() 
plt.show()
```

```python
# Plot the data 
plt.plot(x, y, label="Sales Growth", color='blue') 

# Create a circle 
circle = patches.Circle((3, 25), 10, color='yellow', lw=2, alpha=0.4) 
plt.gca().add_patch(circle) 

# Labels and title 
plt.xlabel("Time (Months)") 
plt.ylabel("Sales ($)") 
plt.title("Sales Growth with Circle") 
plt.legend()
plt.show()
```

```python
# Plot the data 
plt.plot(x, y, label="Sales Growth", color='blue') 

# Create a polygon (triangle in this case) 
polygon = patches.Polygon(((2, 15), (3, 30), (4, 25)), closed=False, color='pink', alpha=0.5) 
plt.gca().add_patch(polygon) 

# Labels and title 
plt.xlabel("Time (Months)") 
plt.ylabel("Sales ($)") 
plt.title("Sales Growth with Polygon") 
plt.legend() 
plt.show() 
```
